```python
from bs4 import BeautifulSoup as bs
import requests
import csv
import pandas as pd
```

```python
link="http://www.nepalstock.com/main/todays_price/index"
```

```python
def scrape_website(link,i,output):
  final_link=f"{link}/{i}"
  url = bs(final_link,'html.parser') 
  soup=requests.get(url)
  print(soup)
  code = soup.text
  lxml_code=bs(code, "lxml")
  table =lxml_code.table
  tag = table.find_all('tr')
  tag[0].text.split('\n')
  for i in range(len(tag)):
      datas = tag[i].text.split('\n')[1:-1]
      if datas[0]!='':
          output.append(datas)
  return output
```

```python
output=[]
for j in range(0,6):
  scrape_website(link,j,output)
```
